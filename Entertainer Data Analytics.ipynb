```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


#load data
basic_info = pd.read_excel("Entertainer - Basic Info.xlsx")
breakthrough_info = pd.read_excel("Entertainer - Breakthrough Info.xlsx")
last_work_info = pd.read_excel("Entertainer - Last work Info.xlsx")
```

```python
#inspect the data
print(basic_info.head())
print(breakthrough_info.head())
print(last_work_info.head())
```

```python
#check the missing values and data types
print(basic_info.info())
print(breakthrough_info.info())
print(last_work_info.info())
```

```python
#data cleaning step

#basic_info:
#convert birth year into numeric, coercing errors to NaN
basic_info["Birth Year"] = pd.to_numeric(basic_info['Birth Year'])

#handle missing values
basic_info.dropna(subset=['Birth Year'], inplace = True)

#remove duplicates
basic_info.drop_duplicates(inplace = True)

basic_info
```

```python
#breakthrough_info:
#convert year column into numeric
breakthrough_info["Year of Breakthrough/#1 Hit/Award Nomination"] = pd.to_numeric(breakthrough_info["Year of Breakthrough/#1 Hit/Award Nomination"])
breakthrough_info["Year of First Oscar/Grammy/Emmy"] = pd.to_numeric(breakthrough_info["Year of First Oscar/Grammy/Emmy"])

#handle missing values
breakthrough_info.dropna(subset = ["Year of First Oscar/Grammy/Emmy"], inplace = True)

#remove duplicate
breakthrough_info.drop_duplicates(inplace= True)

breakthrough_info
```

```python
#last_work_info:
#convert year column into numeric
last_work_info["Year of Last Major Work (arguable)"] = pd.to_numeric(last_work_info["Year of Last Major Work (arguable)"])
last_work_info["Year of Death"] = pd.to_numeric(last_work_info["Year of Death"])

#handle missing values
last_work_info.dropna(subset = ["Year of Death"], inplace = True)

#remove duplicate
last_work_info.drop_duplicates(inplace= True)

last_work_info
```

```python
# Merge datasets on a common key, assuming 'Entertainer Name' is the common column
merged_data = pd.merge(breakthrough_info, last_work_info, on='Entertainer', how='outer')
merged_data = pd.merge(last_work_info, basic_info, on='Entertainer', how='outer')
#merged_data = pd.merge(basic_info,breakthrough_info, on='Entertainer', how='outer')

# Display the merged dataset
print(merged_data.head())
```

```python
# Group by Gender and calculate metrics
gender_analysis = merged_data.groupby('Gender (traditional)').agg({
    'Career Longevity': 'mean'
}).reset_index()

print(gender_analysis)
```

```python
plt.figure(figsize=(3, 6))

# Create the bar plot
ax = sns.barplot(x='Gender (traditional)', y='Career Longevity', data=gender_analysis)

# Set the title and labels
plt.title('Average Career Longevity by Gender')
plt.xlabel('Gender (traditional)')
plt.ylabel('Average Career Longevity (Years)')

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Add the career longevity values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom')

# Show the plot
plt.show()
```
